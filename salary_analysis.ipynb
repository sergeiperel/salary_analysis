```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_excel("salary.xlsx")
df
```

```python
df.rename(columns={"Unnamed: 0":"Отрасль"}, inplace=True )
df
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
years = []
for i in range(24):
    years.append(2000 + i)


mining = df.iloc[1][1:]
construction = df.iloc[2][1:]
education = df.iloc[3][1:]
```

```python
plt.figure(figsize=(15, 7))
plt.plot(years, mining, label = "Добыча полезных ископаемых", marker="o")
plt.plot(years, construction, label = "Строительство", marker="^")
plt.plot(years, education, label = "Образование", marker="s")
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 10000.0))
plt.title("ЗП по отраслям за 2000-2023гг.", fontsize=15)
plt.xlabel("Год", fontsize=15)
plt.ylabel("Зароботная плата", fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
comparison = pd.DataFrame([["Добыча полезных ископаемых", max(mining), min(mining), mining.iloc[23] / mining.iloc[0]],
                  ["Строительство",max(construction), min(construction), construction.iloc[23] / construction.iloc[0]],
                  ["Образование", max(education), min(education), education.iloc[23] / education.iloc[0]]], 
                columns=["Отрасль", "max", "min", "Коэффициент увеличения зарплаты (2023/2000)"])
comparison
```

## Выводы
* Из графика "ЗП по отраслям за 2000-2023гг." видно, что зарплаты по данным трем отраслям экономики за период 2000-2023 все время росли.
* Заработная плата в сфере образования увеличилась в 43 раза, в сфере строительства - в 26 раз, а в сфере полезных ископаемых - в 22 раза.
* В течении всего периода времени средняя зарплата работников отрасли "Добыча полезных ископаемых" выше, чем у работников сферы "Строительство", а средняя зарплата работников отрасли "Строительство" выше, чем у работников сферы "Образование".

```python
inflation = pd.read_excel("inflation.xlsx")
inflation
```

```python
infl = inflation.iloc[0][2:26][::-1]
print(infl[::-1])
print(len(years))
```

```python
plt.figure(figsize=(15, 7))
plt.plot(years, infl, marker="o")

plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(1, 22, 1.0))
plt.title("Динамика роста инфтляции", fontsize=15)
plt.xlabel("Год", fontsize=15)
plt.ylabel("Инфляция", fontsize=15)
plt.grid()
plt.show()
```

```python
data = df.T[1:]
data["Инфляция"] = infl
data = data.rename(columns = {1: "Добыча полезных ископаемых", 2: "Строительство", 3: "Образование"})
data
```

```python
data["Добыча полезных ископаемых с инфляцией"] = data["Добыча полезных ископаемых"] / (1 + data["Инфляция"] / 100)
data["Строительство c инфляцией"] = data["Строительство"] / (1 + data["Инфляция"] / 100)
data["Образование c инфляцией"] = data["Образование"]  / (1 + data["Инфляция"] / 100)
data
```

```python
data.index[0]
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data.index, data["Добыча полезных ископаемых"], label="Номинальные заработные платы", marker="o")
plt.plot(data.index, data["Добыча полезных ископаемых с инфляцией"], label="Реальные заработные платы", marker="^")
plt.title('Динамика номинальных и реальных заработных плат 2000-2023гг. в отрасли "Добыча полезных ископаемых"', fontsize=14)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 10000.0))
plt.xlabel("Год", fontsize=15)
plt.ylabel("Заработанная плата", fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data.index, data["Строительство"], label="Номинальные заработные платы", marker="o")
plt.plot(data.index, data["Строительство c инфляцией"], label="Реальные заработные платы", marker="^")
plt.title('Динамика номинальных и реальных заработных плат 2000-2023гг. в отрасли "Строительство"', fontsize=14)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 10000.0))
plt.xlabel("Год", fontsize=15)
plt.ylabel("Заработанная плата", fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data.index, data["Образование"], label="Номинальные заработные платы", marker="o")
plt.plot(data.index, data["Образование c инфляцией"], label="Реальные заработные платы", marker="^")
plt.title('Динамика номинальных и реальных заработных плат 2000-2023гг. в отрасли "Образование"', fontsize=14)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 10000.0))
plt.xlabel("Год", fontsize=15)
plt.ylabel("Заработанная плата", fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data.index, data["Добыча полезных ископаемых"], label="Добыча полезных ископаемых (номинальные зп)", marker="o")
plt.plot(data.index, data["Добыча полезных ископаемых с инфляцией"], label="Добыча полезных ископаемых (реальные зп)", marker="^", linestyle="--")
plt.plot(data.index, data["Строительство"], label="Строительство (номинальные зп)", marker="o")
plt.plot(data.index, data["Строительство c инфляцией"], label="Строительство (реальные зп)", marker="^", linestyle="--")
plt.plot(data.index, data["Образование"], label="Образование (номинальные зп)", marker="o")
plt.plot(data.index, data["Образование c инфляцией"], label="Образование (реальные зп)", marker="^", linestyle="--")
plt.title("Отрасли экономики с учетом инфляции и без", fontsize=14)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(2000, 150000, 10000.0))
plt.xlabel('Год', fontsize=15)
plt.ylabel('Заработанная плата', fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

```python
data_change = data
```

```python
data_change["Добыча полезных ископаемых с инфляцией (пред.)"] = data["Добыча полезных ископаемых с инфляцией"].shift(1)
data_change["Строительство c инфляцией (пред.)"] = data["Строительство c инфляцией"].shift(1)
data_change["Образование c инфляцией (пред.)"] = data["Образование c инфляцией"].shift(1)
data_change
```

```python
data_change["Влияние инфляции на Добычу полезных ископаемых"] = (data_change["Добыча полезных ископаемых с инфляцией"] / data_change["Добыча полезных ископаемых с инфляцией (пред.)"] - 1) * 100
data_change["Влияние инфляции на Строительство"] = (data_change["Строительство c инфляцией"] / data_change["Строительство c инфляцией (пред.)"] - 1) * 100
data_change["Влияние инфляции на Образование"] = (data_change["Образование c инфляцией"] / data_change["Образование c инфляцией (пред.)"] - 1) * 100
data_change
```

```python
plt.figure(figsize=(15, 7))
plt.plot(data_change.index, data_change["Влияние инфляции на Добычу полезных ископаемых"], label="Добыча полезных ископаемых", marker="o")
plt.plot(data_change.index, data_change["Влияние инфляции на Строительство"], label="Строительство", marker="^")
plt.plot(data_change.index, data_change["Влияние инфляции на Образование"], label="Образование", marker="s")
plt.plot(data_change.index, data_change["Инфляция"], label="Годовая инфляция", marker="+")

plt.title("Влияние инфляции на изменение зарплаты по сравнению с предыдущим годом в период 2000-2023гг.", fontsize=14)
plt.xticks(np.arange(2000, 2024, 1.0))
plt.yticks(np.arange(0, 70, 5.0))
plt.xlabel('Год', fontsize=15)
plt.ylabel('Заработанная плата', fontsize=15)
plt.legend(fontsize=15)
plt.grid()
plt.show()
```

## Выводы:
* Практически во все годы заработные платы опережают темпы инфляции.
* В некоторые годы, а именно в 2009, 2010, 2014, 2015, 2020, 2022 годы можно заметить, что темпы инфляции опережают изменения в заработных платах.
* Данные наблюдения можно объяснить различными кризисами, нестабильной ситуацией в стране и мире или другими экономическими факторами.

## Выводы
* Из корреляцилнной матрицы видно, что существует тесная связь между зарплатами в России и уровнем ВВП, что говорит о том, что экономический рост влияет на увеличение доходов работников всех отраслей экономики.
* Те же наблюдения можно констатировать и для зарплат с учетом инфляции.
* Также наблюдается обратно пропорциональная зависимость между годовой инфляцией и всеми зарплатами по отраслям, а также годовой инфляцией и ВВП. Это означает, что инфляция растет, а зарплаты снижатся. Инфляция растет, а ВВП падает.
* Также видна положительная корреляция между показателями инфляции и влиянием инфляции на изменение заработных плат по сравнению с предыдущим годом.
